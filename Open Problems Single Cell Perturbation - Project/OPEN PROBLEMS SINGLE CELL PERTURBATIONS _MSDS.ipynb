## Exploratory Data Analysis (EDA) - Inspecting, Visualizing, and Cleaning the Data [26 points]

### 1. Dataset Component Description
In this step, we will describe each factor or feature that constitutes the dataset. For each factor, various visualization techniques such as box-plots, scatter plots, and histograms will be employed to describe the data distribution. This will give us a clear understanding of the range, central tendency, and dispersion of each feature in the dataset.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the de_train.csv dataset

de_train_df = pd.read_parquet(de_train_path)

# Identify the top 10 compounds (sm_name) in the dataset
top_sm_name = de_train_df['sm_name'].value_counts().index[:10]


# Identify gene columns that contain valid numeric data
valid_gene_columns = []
for gene in de_train_df.columns:
    if gene not in ['cell_type', 'sm_name', 'sm_lincs_id', 'SMILES', 'control']:
        gene_data = pd.to_numeric(de_train_df[gene], errors='coerce')
        valid_gene_data = gene_data.dropna()
        if not valid_gene_data.empty and np.issubdtype(valid_gene_data.dtype, np.number):
            valid_gene_columns.append(gene)
            if len(valid_gene_columns) == 2:
                break

# Set up the matplotlib figure
fig, axes = plt.subplots(3, 2, figsize=(15, 12))
fig.suptitle('Distribution of Features in de_train.csv')

# Feature 'cell_type' distribution
sns.countplot(ax=axes[0, 0], y='cell_type', data=de_train_df, order=de_train_df['cell_type'].value_counts().index)
axes[0, 0].set_title('Distribution of Cell Types')
axes[0, 0].set_xlabel('Count')
axes[0, 0].set_ylabel('Cell Type')

# Feature 'sm_name' distribution (top 10 compounds)
sns.countplot(ax=axes[0, 1], y='sm_name', data=de_train_df[de_train_df['sm_name'].isin(top_sm_name)], order=top_sm_name)
axes[0, 1].set_title('Distribution of Compounds (Top 10)')
axes[0, 1].set_xlabel('Count')
axes[0, 1].set_ylabel('Compound (sm_name)')

# Feature 'control' distribution
sns.countplot(ax=axes[1, 0], x='control', data=de_train_df)
axes[1, 0].set_title('Distribution of Controls')
axes[1, 0].set_xlabel('Control')
axes[1, 0].set_ylabel('Count')

# Histogram of expression levels for valid selected genes
for i, gene in enumerate(valid_gene_columns):
    gene_data = pd.to_numeric(de_train_df[gene], errors='coerce')
    valid_gene_data = gene_data.dropna()
    sns.histplot(ax=axes[1 + i, 1], x=valid_gene_data, bins=30, kde=True)
    axes[1 + i, 1].set_title(f'Distribution of {gene}')
    axes[1 + i, 1].set_xlabel('Expression Level')
    axes[1 + i, 1].set_ylabel('Density')

# Remove the empty subplot
fig.delaxes(axes[2, 0])

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```


### 2. Correlation Analysis
We will analyze the correlations between different factors in the dataset. This involves both numerical analysis, using correlation coefficients, and graphical analysis, using heatmaps or scatter plots, to justify whether the factors are correlated or not. Understanding the correlations will help in identifying the relationships between different features and their potential impact on the predictive models.



```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'RAW_DATA/de_train.parquet'
de_train_df = pd.read_parquet(file_path)

# Extract gene expression columns
gene_columns = de_train_df.columns.difference(['cell_type', 'sm_name', 'sm_lincs_id', 'SMILES', 'control'])

# Define the block size
block_size = 1000

# Initialize a list to store the top correlations from each block
top_correlations_list = []

# Loop over the blocks of genes to calculate correlation and extract top correlations
num_blocks = len(gene_columns) // block_size + 1
for i in range(num_blocks):
    for j in range(i, num_blocks):
        # Define the start and end index for each block
        start_idx_i = i * block_size
        end_idx_i = min((i + 1) * block_size, len(gene_columns))
        start_idx_j = j * block_size
        end_idx_j = min((j + 1) * block_size, len(gene_columns))
        
        # Select the genes in the current block
        block_genes_i = gene_columns[start_idx_i:end_idx_i]
        block_genes_j = gene_columns[start_idx_j:end_idx_j]
        
        # Calculate the correlation matrix for the current block
        block_correlation_matrix = de_train_df[block_genes_i].corrwith(de_train_df[block_genes_j])
        
        # Convert the correlation matrix to long format and add to the list
        for gene1, correlation in block_correlation_matrix.items():
            top_correlations_list.append({'Gene1': gene1, 'Gene2': block_genes_j[0], 'Correlation': correlation})

# Create a DataFrame from the list
top_correlations_per_block = pd.DataFrame(top_correlations_list)

# Calculate the absolute value of correlation and sort the DataFrame by it
top_correlations_per_block['AbsoluteCorrelation'] = top_correlations_per_block['Correlation'].abs()
top_correlations_per_block = top_correlations_per_block.sort_values(by='AbsoluteCorrelation', ascending=False)

# Select the overall top 50 strongest correlations
overall_top_50_correlations = top_correlations_per_block.drop_duplicates(subset=['Gene1', 'Gene2']).head(50)

# Display the overall top 50 strongest correlations
print(overall_top_50_correlations.head())

# Create a heat map of the overall top 50 strongest correlations
plt.figure(figsize=(14, 12))
sns.heatmap(overall_top_50_correlations.pivot(index='Gene1', columns='Gene2', values='Correlation'), cmap='coolwarm', annot=True)
plt.title('Heatmap of Overall Top 50 Strongest Correlations')
plt.show()
```

```python
import numpy as np

# Select only numeric columns
numeric_cols = de_train_df.select_dtypes(include=np.number).columns

# Identify outliers using the IQR method
Q1 = de_train_df[numeric_cols].quantile(0.25)
Q3 = de_train_df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for the outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find the outliers
outliers = ((de_train_df[numeric_cols] < lower_bound) | (de_train_df[numeric_cols] > upper_bound))

# Display the number of identified outliers
print(f"Number of outliers identified:\n{outliers.sum()}")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the standardized and log-transformed dataset
file_path = 'TRANSFORM_DATA/de_train_transformed.csv'
df = pd.read_csv(file_path)

# Identify gene expression columns
gene_columns = df.columns.difference(['cell_type', 'sm_name', 'sm_lincs_id', 'SMILES', 'control'])

# Calculate the IQR for each gene expression column
Q1 = df[gene_columns].quantile(0.25)
Q3 = df[gene_columns].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Visualize the distribution of gene expression data before imputation
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(df[gene_columns].values.flatten(), bins=50, kde=True)
plt.title('Distribution Before Imputation')
plt.xlabel('Gene Expression')
plt.ylabel('Frequency')

# Impute outliers with the median of each column
df[gene_columns] = np.where(df[gene_columns] < lower_bound, df[gene_columns].median(), df[gene_columns])
df[gene_columns] = np.where(df[gene_columns] > upper_bound, df[gene_columns].median(), df[gene_columns])

# Visualize the distribution of gene expression data after imputation
plt.subplot(1, 2, 2)
sns.histplot(df[gene_columns].values.flatten(), bins=50, kde=True)
plt.title('Distribution After Imputation')
plt.xlabel('Gene Expression')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Save the modified dataset to a CSV file
output_file_path = 'TRANSFORM_DATA/de_train_standardized_imputed.csv'
df.to_csv(output_file_path, index=False)
```

   - **Dimensionality Reduction Models**: such as PCA or t-SNE for visualizing high-dimensional data in a lower-dimensional space.
   



```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# Set the number of OpenBLAS threads to 1
os.environ['OPENBLAS_NUM_THREADS'] = '1'

# Load the standardized and imputed dataset
file_path = 'TRANSFORM_DATA/de_train_standardized_imputed.csv'
data_df = pd.read_csv(file_path, index_col=0)

# Extract gene expression columns for model building
gene_columns = data_df.columns.difference(['cell_type', 'sm_name', 'sm_lincs_id', 'SMILES', 'control'])
X = data_df[gene_columns]

# K-means: Elbow method to determine the optimal number of clusters
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
    
# Plotting the Elbow method graph
plt.figure(figsize=(10,6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()

# Based on the elbow plot, choose an appropriate value for k and build the K-means model
optimal_k = 4  # Replace with the chosen value of k
kmeans = KMeans(n_clusters=optimal_k, n_init=10, random_state=42)
data_df['Cluster'] = kmeans.fit_predict(X)

# PCA for dimensionality reduction and visualization
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(X)
principalDf = pd.DataFrame(data=principalComponents, columns=['Principal Component 1', 'Principal Component 2'])

# Visualizing 2D Projection using PCA
plt.figure(figsize=(10,6))
plt.scatter(principalDf['Principal Component 1'], principalDf['Principal Component 2'], c=data_df['Cluster'])
plt.title('2D PCA Projection')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

# t-SNE for visualization
tsne = TSNE(n_components=2, random_state=42)
tsne_results = tsne.fit_transform(X)

# Visualizing using t-SNE
fig, ax = plt.subplots(figsize=(10,6))  # Criando uma figura e um eixo
scatter = ax.scatter(tsne_results[:, 0], tsne_results[:, 1], c=data_df['Cluster'])

# Adding a legend
legend1 = ax.legend(*scatter.legend_elements(), loc="upper right", title="Clusters")
ax.add_artist(legend1)

plt.title('t-SNE Visualization')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.show()

```

### 2. Hyperparameter Optimization
A critical aspect of building robust models is hyperparameter optimization. We will employ techniques such as grid search or random search to find the optimal set of hyperparameters for each model, enhancing their performance and predictive accuracy.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score
from itertools import product

# Carregar os dados
data_df = pd.read_csv('TRANSFORM_DATA/de_train_standardized_imputed.csv')
gene_columns = data_df.columns.difference(['cell_type', 'sm_name', 'sm_lincs_id', 'SMILES', 'control'])
X = data_df[gene_columns]

# Inicializar o DataFrame de resultados
results_df = pd.DataFrame(columns=['Algorithm', 'Parameters', 'Silhouette Score'])

# Busca em grade para KMeans
kmeans_params = {
    'n_clusters': list(range(2, 11)),
    'init': ['k-means++', 'random'],
    'n_init': [10]
}
for params in product(*kmeans_params.values()):
    model = KMeans(**dict(zip(kmeans_params.keys(), params)))
    labels = model.fit_predict(X)
    score = silhouette_score(X, labels)
    new_row = pd.DataFrame({'Algorithm': ['KMeans'], 'Parameters': [params], 'Silhouette Score': [score]})
    results_df = pd.concat([results_df, new_row], ignore_index=True)

# Busca em grade para DBSCAN
dbscan_params = {
    'eps': [0.1, 0.5, 1, 1.5, 2],
    'min_samples': list(range(5, 21))
}
for params in product(*dbscan_params.values()):
    model = DBSCAN(**dict(zip(dbscan_params.keys(), params)))
    labels = model.fit_predict(X)
    if len(set(labels)) > 1:
        score = silhouette_score(X, labels)
        new_row = pd.DataFrame({'Algorithm': ['DBSCAN'], 'Parameters': [params], 'Silhouette Score': [score]})
        results_df = pd.concat([results_df, new_row], ignore_index=True)

# Busca em grade para Agglomerative Clustering
agg_params = {
    'n_clusters': list(range(2, 11)),
    'linkage': ['single', 'complete', 'average', 'ward']
}
for params in product(*agg_params.values()):
    model = AgglomerativeClustering(**dict(zip(agg_params.keys(), params)))
    labels = model.fit_predict(X)
    score = silhouette_score(X, labels)
    new_row = pd.DataFrame({'Algorithm': ['AgglomerativeClustering'], 'Parameters': [params], 'Silhouette Score': [score]})
    results_df = pd.concat([results_df, new_row], ignore_index=True)

# Ordenar os resultados pela pontuaÃ§Ã£o do Silhouette Score em ordem decrescente
results_df = results_df.sort_values(by="Silhouette Score", ascending=False)

# Criar labels para o eixo Y combinando o nome do algoritmo e os parÃ¢metros
results_df["Algorithm + Parameters"] = results_df["Algorithm"] + "\n" + results_df["Parameters"].astype(str)

# Selecionar os 10 melhores resultados
top_10_results_df = results_df.head(10)

# Criar um grÃ¡fico de barras horizontal com Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Silhouette Score', y='Algorithm + Parameters', data=top_10_results_df, palette='viridis')
plt.title('Top 10 Silhouette Scores for Clustering Algorithms')
plt.show()

```

### 3. Model Training and Validation
Each selected model will be trained using the cleaned and preprocessed data from the EDA phase. The validation will involve assessing the models' performance using appropriate metrics, such as silhouette score for clustering models or explained variance ratio for dimensionality reduction models.



```python
# Obter o melhor conjunto de parÃ¢metros do DataFrame de resultados
best_result = top_10_results_df.iloc[0]
best_algorithm = best_result['Algorithm']
best_parameters_tuple = best_result['Parameters']

# Converter a tupla de parÃ¢metros em um dicionÃ¡rio
if best_algorithm == 'KMeans':
    best_parameters = dict(zip(['n_clusters', 'init', 'n_init'], best_parameters_tuple))
elif best_algorithm == 'DBSCAN':
    best_parameters = dict(zip(['eps', 'min_samples'], best_parameters_tuple))
elif best_algorithm == 'AgglomerativeClustering':
    best_parameters = dict(zip(['n_clusters', 'linkage'], best_parameters_tuple))

# Treinar o modelo com o melhor conjunto de parÃ¢metros
if best_algorithm == 'KMeans':
    model = KMeans(**best_parameters)
elif best_algorithm == 'DBSCAN':
    model = DBSCAN(**best_parameters)
elif best_algorithm == 'AgglomerativeClustering':
    model = AgglomerativeClustering(**best_parameters)

# Ajustar o modelo aos dados
labels = model.fit_predict(X)

# Calcular a pontuaÃ§Ã£o Silhouette para o modelo treinado
score = silhouette_score(X, labels)

# Adicionar os rÃ³tulos dos grupos ao DataFrame para anÃ¡lise
data_df['Cluster'] = labels

# Por exemplo, a contagem de amostras em cada grupo
cluster_sizes = data_df.groupby('Cluster').size()

# Exibir os resultados
print(f"O melhor modelo Ã© {best_algorithm} com parÃ¢metros {best_parameters}")
print(f"PontuaÃ§Ã£o Silhouette para o melhor modelo: {score}")
print("Tamanho dos Grupos:\n", cluster_sizes)
```

### 4. Comparative Analysis

In this section, we will conduct a comparative analysis to understand the strengths and weaknesses of each model, namely K-Means, DBSCAN, and Agglomerative Clustering. We will evaluate their performance metrics, analyze the results, and discuss the implications of our findings.

#### 4.1. Performance Metrics
- **Agglomerative Clustering Model:**
   - Achieved the highest Silhouette score of 0.0795.
   - The Silhouette score ranges from -1 to 1. A high score indicates that the object is well matched to its own cluster and poorly matched to neighboring clusters.

#### 4.2. Cluster Visualization
- **t-SNE Visualizations:**
   - Depict the formation of clusters in the reduced-dimensional space.
   - The majority of the data points (607) are contained in one cluster, while the others contain very few points (1 to 4).

#### 4.3. Analysis of Models
   - **a) K-Means:**
      - **Strengths:**
         - Easy to implement and interpret.
         - Computationally efficient.
      - **Weaknesses:**
         - Sensitive to initialization.
         - Assumes spherical and equally sized clusters.
      - **Cautions:**
         - The choice of the number of clusters $k$ is critical.
         - Sensitive to outliers.

   - **b) DBSCAN:**
      - **Strengths:**
         - Can find arbitrarily shaped clusters.
         - Doesn't require specifying the number of clusters.
      - **Weaknesses:**
         - Sensitive to the choice of `eps` and `min_samples` parameters.
         - May struggle with varying cluster densities.
      - **Cautions:**
         - Proper parameter selection is crucial.
         - May not perform well with clusters of different densities.

   - **c) Agglomerative Clustering:**
      - **Strengths:**
         - Flexibility with different linkage metrics.
         - Can produce a hierarchy of clusters.
      - **Weaknesses:**
         - Higher computational cost for large datasets.
      - **Cautions:**
         - The choice of linkage criterion affects the shape of the clusters.
         - The choice of the number of clusters is still required.


#### 4.4. Conclusions
   - The Agglomerative Clustering model achieved the best Silhouette score, but the uneven distribution of cluster sizes is noteworthy.
   - Each model has its unique strengths and weaknesses; the choice of model depends on the context and specific requirements of the problem.
   - Visual analysis and interpretation of the clusters are essential to validate the results and extract meaningful insights.


#### 4.5. Recommendations
   - For large datasets, more efficient models such as K-Means or DBSCAN may be preferable.
   - Experimentation with different parameters and preprocessing techniques may lead to enhanced results.
   - A deeper analysis of the formed clusters and validation with domain knowledge are necessary to confirm the relevance of the found groupings.

Upon reviewing this analysis, the main considerations are the model's suitability to the data type, the interpretability of the results, and the practical relevance of the identified clusters.
